```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
M = 10000

amp = np.random.random(N) * 5

fre = np.random.random(N) * 100

pha = np.random.random(N) * 2*np.pi
```

```python
amp.size
```

```python
amp
```

```python
dat = np.zeros(M)
for I in range(N):
    dat += amp[I]*np.sin(2*np.pi*fre[I]*np.arange(0,M,1)+pha[I])
```

```python
plt.plot(dat)
```
