```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
x = np.arange(0,50000,0.01)*np.pi
y = np.sin(x) + 7*np.sin(2.3*x) + 3*np.sin(20*x) + 20*np.random.normal(0,1,x.size)
#plt.plot(x,y)

N = x.size
# sample spacing

yf = np.abs(fft(y)[:N//2])
xf = fftfreq(N,x[1])[:N//2]

plt.plot(xf,yf)
plt.xlim(0,5)
plt.yscale('log')
```

```python
import numpy as np
def AutoCorrelation(x):
    y = x-x.mean()
    result = np.correlate(y, y, "full")/np.std(y)**2
    return result 
```
