```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10

M = 10000

data = np.zeros(M)
```

```python
amp = np.random.random(N) * 5

fre = np.random.random(N)

pha = np.random.random(N) * (2*np.pi)
```

```python
for i in range(N):
    data += amp[i] * np.sin(2*np.pi*fre[i]*np.arange(0,M,1)+pha[i])
```

```python
data
```

```python
nois = (2*np.random.random(M)-1)*3
snr = data.sum()/nois.sum()
data += nois
```

```python
plt.plot(data)
```

```python
from scipy.fft import fft, fftfreq
```

```python
yf = np.abs(fft(data)[:M//2])
xf = fftfreq(M,1)[:M//2]

plt.plot(xf,yf)
#plt.xlim(0,5)
plt.yscale('log')
plt.text(0.4,1e4,'SNR = {0:.2f}'.format(snr))
```
