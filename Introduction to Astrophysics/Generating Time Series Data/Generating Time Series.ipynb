```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
N = 1000

data = np.zeros(N)
```

```python
fre = np.random.random(N)
#freq = 1**fre
pow_ind = -2
amp = np.zeros(N)
#const = 0.01

noise = (50*np.random.random(N)-1)*3


def Plaw(amp, fre, pow_ind, noise):
    amp += 50*fre** pow_ind
    return amp

pha = np.random.random(N) * (2*np.pi)

T = np.arange(N)
```

```python
Plaw(amp, fre, pow_ind, noise)
```

```python
for i in range(N):
    data += amp[i] * np.sin(2*np.pi*fre[i]* T )
```

```python
nois = (2*np.random.random(N)-1)*3
#snr = data.sum()/nois.sum()
data += nois
```
